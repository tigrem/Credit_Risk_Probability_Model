```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_style("whitegrid")
# To ensure plots 
%matplotlib inline

print("Libraries loaded successfully.")
# --- Load Data ---
data_path = '../data/raw/data.csv'
try:
    df = pd.read_csv(data_path)
    print(f"Data loaded successfully from '{data_path}'")
except FileNotFoundError:
    print(f"Error: {data_path} not found. Make sure it's in the 'data/raw/' directory.")
    print("Please place 'data.csv'")
    raise FileNotFoundError(f"Data file not found at {data_path}")

# Overview of the Data
print("\n--- Data Overview ---")
print("Shape of the dataset (rows, columns):", df.shape)
print("\nFirst 5 rows of the dataset:")
print(df.head())
print("\nInformation about columns (data types, non-null counts):")
df.info()

#Summary Statistics
print("\n--- Summary Statistics for Numerical Features ---")
print(df.describe())
# Check for numerical columns
for col in df.select_dtypes(include=np.number).columns:
    if df[col].min() == df[col].max():
        print(f"Warning: Numerical column '{col}' has a constant value ({df[col].min()}). It will not show variation in plots.")

print("\n--- Summary Statistics for Categorical Features ---")
print(df.describe(include='object'))


#Distribution of Numerical Features
print("\n--- Distribution of Numerical Features ---")
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

if not numerical_cols:
    print("No numerical columns found to plot distributions.")
else:
    # Adjust figure size
    num_plots = len(numerical_cols)
    cols_per_row = 3
    rows_needed = (num_plots + cols_per_row - 1) // cols_per_row 
    plt.figure(figsize=(18, rows_needed * 5)) 

    for i, col in enumerate(numerical_cols):
        plt.subplot(rows_needed, cols_per_row, i + 1)
        sns.histplot(df[col], kde=True, bins=50) # Increased bins for more detail
        plt.title(f'Distribution of {col}', fontsize=14)
        plt.xlabel(col, fontsize=12)
        plt.ylabel('Frequency', fontsize=12)
        # Add text for min/max/mean
        plt.text(0.05, 0.95, f'Min: {df[col].min():.2f}', transform=plt.gca().transAxes, fontsize=10, verticalalignment='top')
        plt.text(0.05, 0.90, f'Max: {df[col].max():.2f}', transform=plt.gca().transAxes, fontsize=10, verticalalignment='top')
        plt.text(0.05, 0.85, f'Mean: {df[col].mean():.2f}', transform=plt.gca().transAxes, fontsize=10, verticalalignment='top')
    plt.tight_layout(pad=3.0) # Added padding 
    plt.show()

#Distribution of Categorical Features ---
print("\n--- Distribution of Categorical Features ---")
categorical_cols = df.select_dtypes(include='object').columns.tolist()

if not categorical_cols:
    print("No categorical columns found to plot distributions.")
else:
    num_plots = len(categorical_cols)
    cols_per_row = 2
    rows_needed = (num_plots + cols_per_row - 1) // cols_per_row
    plt.figure(figsize=(16, rows_needed * 6)) # Increased figure size

    for i, col in enumerate(categorical_cols):
        plt.subplot(rows_needed, cols_per_row, i + 1)
        # Limit to top N categories
        if df[col].nunique() > 20:
            top_categories = df[col].value_counts().head(20).index
            sns.countplot(y=df[col], order=top_categories, palette='viridis')
            plt.title(f'Top 20 Count of {col}', fontsize=14)
        else:
            sns.countplot(y=df[col], order = df[col].value_counts().index, palette='viridis')
            plt.title(f'Count of {col}', fontsize=14)
        plt.xlabel('Count', fontsize=12)
        plt.ylabel(col, fontsize=12)
    plt.tight_layout(pad=3.0)
    plt.show() # Display the plots

#Correlation Analysis ---
print("\n--- Correlation Matrix of Numerical Features ---")
if not numerical_cols:
    print("No numerical columns for correlation analysis.")
else:
    correlation_matrix = df[numerical_cols].corr()
    print(correlation_matrix)

    plt.figure(figsize=(14, 12)) # Increased figure size
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, annot_kws={"size": 10})
    plt.title('Correlation Matrix of Numerical Features', fontsize=16)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10, rotation=0)
    plt.show() # Display the plot

# Identifying Missing Values ---
print("\n--- Missing Values Check ---")
missing_values = df.isnull().sum()
missing_values_percent = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percent': missing_values_percent
})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False)
print(missing_df)

if missing_df.empty:
    print("No missing values found in the dataset.")
else:
    print("\nSummary of Columns with Missing Values:")
    print(missing_df)

# Outlier Detection (using Box Plots)
print("\n--- Outlier Detection (Box Plots for Numerical Features) ---")
if not numerical_cols:
    print("No numerical columns to plot for outlier detection.")
else:
    num_plots = len(numerical_cols)
    cols_per_row = 3
    rows_needed = (num_plots + cols_per_row - 1) // cols_per_row
    plt.figure(figsize=(18, rows_needed * 5)) # Increased figure size

    for i, col in enumerate(numerical_cols):
        plt.subplot(rows_needed, cols_per_row, i + 1)
        sns.boxplot(y=df[col], palette='Pastel1')
        plt.title(f'Box Plot of {col}', fontsize=14)
        plt.ylabel(col, fontsize=12)
    plt.tight_layout(pad=3.0)
    plt.show() # Display the plots

print("\n--- EDA  task completed ---")

```
